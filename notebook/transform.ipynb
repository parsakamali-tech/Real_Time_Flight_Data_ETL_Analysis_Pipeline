```python
import pandas as pd
import numpy as np
from pyod.models.iforest import IForest
```

```python
data = pd.read_csv('..//database//api_data//api_data.csv')
data
```

```python
data.columns
```

```python
null = {c:int(data[c].isnull().sum()) for c in data.columns}
null
```

```python
data.drop([c for c in null if null[c] == 100], axis=1,inplace=True)
data
```

```python
data.columns
```

```python
data.drop_duplicates(inplace=True)
data
```

------------------------------------------

Date Formating

```python
datetime_columns = [
    'departure_scheduled',
    'departure_estimated',
    'departure_actual',
    'departure_estimated_runway',
    'departure_actual_runway',
    'arrival_scheduled',
    'flight_date'
    ]
data[datetime_columns] = data[datetime_columns].apply(pd.to_datetime, errors='coerce')
```

Calculating flight duration from the difference between arrival and departure

```python
data['flight_duration'] = np.where(
    data['arrival_scheduled'] > data['departure_scheduled'],# condition of duration
    (data['arrival_scheduled'] - data['departure_scheduled']).dt.total_seconds()/3600, # true
    (data['departure_scheduled'] - data['arrival_scheduled']).dt.total_seconds()/3600 # false
)
data['flight_duration'] = data['flight_duration'].fillna(data['flight_duration'].mean())
```

-------------------------------------

Filiing null values for Status columns and IDs

```python
column = 'flight_status'
data[column] = data[column].fillna("unknown")
data
```

Filiing null values for Origin/Destination Airport Columns

```python
columns = ['departure_terminal','arrival_terminal','departure_gate','arrival_gate','arrival_baggage']
data[columns] = data[columns].fillna("unknown")
data
```

Filling null values for Delays

```python
column = 'departure_delay'
data[column] = data[column].fillna(0) # i assumed that null means 0 beacause there is a lot of null field
data
```

Filling null values for departure_actual when active --> with departure_scheduled and when --> scheduled with 'null'

```python
column = 'departure_actual'
data[column] = np.where(
    data['flight_status'] == 'active',
    data[column].fillna(data['departure_scheduled']),
    np.where(
        data['flight_status'] == 'scheduled',
        data[column].fillna('null'),
        data[column]
    )
)
data[column].head(20)
```

Filling null values for departure_estimated_runway & departure_actual_runway when active --> with departure_actual and when --> scheduled with 'null'

```python
column = 'departure_estimated_runway'
data[column] = np.where(
    data['flight_status'] == 'active',
    data[column].fillna(data['departure_actual']),
    np.where(
        data['flight_status'] == 'scheduled',
        data[column].fillna('null'),
        data[column]
    )
)
column5 = 'departure_actual_runway'
data[column5] = np.where(
    data['flight_status'] == 'active',
    data[column5].fillna(data['departure_actual']),
    np.where(
        data['flight_status'] == 'scheduled',
        data[column5].fillna('null'),
        data[column5]
    )
)
```

Filling null values for flight codeshared columns

```python
columns = ['flight_codeshared_airline_name',
           'flight_codeshared_airline_iata',
           'flight_codeshared_airline_icao',
           'flight_codeshared_flight_number',
           'flight_codeshared_flight_iata',
           'flight_codeshared_flight_icao']
data[columns] = data[columns].fillna('N/A')
data
```

Filling null values for arrival_timezone with destination airport code (iata)

```python
column = 'arrival_timezone'
data[column] = data[column].fillna(data['arrival_iata'])
```

Filling null values for 'flight_iata','flight_icao','flight_number' by flight_number if both flight_iata,flight_icao are null

```python
columns = ['flight_iata','flight_icao']
column1 = 'flight_number'
column2 = 'flight_iata'
column3 = 'flight_icao'

data[column1] = data[column1].fillna(0)
data[column2] = data[column2].fillna(data[column3])
data[column3] = data[column3].fillna(data[column2])

n = data[column2].isnull().sum() + data[column3].isnull().sum()
if n != 0 :
    data[column2] = data[column2].fillna(data['flight_number'])
    data[column3] = data[column3].fillna(data['flight_number'])
```

Checking null leftovers

```python
null = {c:int(data[c].isnull().sum()) for c in data.columns}
null
```

Filling rest of null values

```python
column = 'arrival_airport'
data[column] = data[column].fillna(data['flight_number'])
```

```python
column2 = 'airline_iata'
column3 = 'airline_icao'

data[column2] = data[column2].fillna(data[column3])
data[column3] = data[column3].fillna(data[column2])
n = data[column2].isnull().sum() + data[column3].isnull().sum()
if n != 0 :
    data[column2] = data[column2].fillna(data['airline_name'])
    data[column3] = data[column3].fillna(data['airline_name'])
```

-------------------------------------------------------------------

------------------------------------------------------------------

Using pyOD-->IForest model for understanding departure_delay outliers in outlier column

```python
column = 'departure_delay'
model = IForest()
model.fit(data[[column]].values)
outlier_labels = model.predict(data[[column]].values)
data['outlier'] = outlier_labels
data['outlier'] = data['outlier'].astype(bool)
```

```python
column = 'flight_duration'
model = IForest()
model.fit(data[[column]].values)
outlier_labels = model.predict(data[[column]].values)
data['flight_duration_outlier'] = outlier_labels
data['flight_duration_outlier'] = data['flight_duration_outlier'].astype(bool)
```

```python
data = data.rename(columns={'outlier':'delay_outlier'})
```

```python
data.to_csv("..//database//transformed_data//transformed_data.csv", encoding="utf-8" ,index=False)
```
